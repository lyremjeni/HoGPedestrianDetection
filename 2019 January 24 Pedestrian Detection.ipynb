```python
"""
NICE PARAMETERS: 
1. frame size: (320, 240), stride: (6,6), padding: (20,20), scale=1.05 
2. frame size: B&W (320, 240), stride: (8,8), padding: (8,8), scale=1.05 
3. frame size: B&W (320, 240), stride: (4, 4), padding: (8,8), scale=1.05 real-time but a bit false positives
4. 
"""

import cv2
import time
from imutils.object_detection import non_max_suppression
import numpy as np

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor.getDefaultPeopleDetector())
cap = cv2.VideoCapture(0)
frame_scale = 1
try:
    while True:
        r, frame_orig = cap.read()
        if r:
            start_time = time.time()
            frame = cv2.resize(frame_orig,(320, 240)) # Downscale to improve frame rate
            gray_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY) # HOG needs a grayscale image

    #         rects, weights = hog.detectMultiScale(gray_frame, winStride=(4, 4), scale=1.0)
            rects, weights = hog.detectMultiScale(cv2.UMat(gray_frame), winStride=(8, 8), padding=(8, 8), scale=1.05)

    #         Measure elapsed time for detections
            end_time = time.time()
    #         print("Elapsed time:", end_time-start_time)

            for i, (x, y, w, h) in enumerate(rects):
                print(x,y,weights[i])
                if weights[i] < 0.7:
                    continue
                
                cv2.rectangle(frame, (x,y), (x+w,y+h),(0,255,0),2)

            rects = np.array([[frame_scale*x, frame_scale*y, frame_scale*(x + w), frame_scale*(y + h)] for (x, y, w, h) in rects])
            pick = non_max_suppression(rects, probs=None, overlapThresh=0.25)

            for (xA, yA, xB, yB) in pick:
                cv2.rectangle(frame_orig, (xA, yA), (xB, yB), (0, 255, 0), 2)


            cv2.imshow("preview 2", frame)

        k = cv2.waitKey(1)
        if k & 0xFF == ord("q"): # Exit condition
            break
except:
    cap.release()
            
"""
January 23, 2019
"""
```
